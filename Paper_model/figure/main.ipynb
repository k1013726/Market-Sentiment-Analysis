```python
import matplotlib.pyplot as plt
import pandas as pd
import requests

df=pd.read_csv('../Data/BTC-all.csv')
df["Date"] = pd.to_datetime(df["Date"])


plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Close"], label="Close", color="b")

plt.xlabel("Date")
plt.ylabel("Close(USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
